```python
# bai 2
class Ward:
    def __init__(self,name):
        self.ward_name = name
        self.people = []
        
    def addPerson(self,person):
        self.people.append(person)
        
    def describe(self):
        print(f"Ward name: {self.ward_name}")
        for person in self.people:
            person.describe()
        
    def countDoctor(self):
        self.count = 0
        for person in (self.people):
            if isinstance(person,Doctor):
                self.count +=1
        return self.count
    
    def sortAge(self):
        sort_list =sorted(self.people,key = lambda person: person.yob, reverse=True)
        return sort_list
    
    def aveTeacherYearOfBirth(self):
        total_age = 0
        for i,person in enumerate(self.people):
            total_age += int(person.yob)
            average = total_age/ (i+1)
        print(average)
    
    class Student:
        def __init__(self, name,yob,grade):
            self.name = name
            self.yob = yob
            self.grade = grade
        def describe(self):
            print(f"Student - Name:{self.name} - Yob:{self.yob} - Grade:{self.grade}")
            
    class Teacher:
        def __init__(self, name,yob,subject):
            self.name = name
            self.yob = yob
            self.subject = subject
        def describe(self):
            print(f"Teacher - Name:{self.name} - Yob:{self.yob} - Subject:{self.subject}")
            
    class Doctor:
        def __init__(self, name,yob,specialist):
            self.name = name
            self.yob = yob
            self.specialist = specialist
        def describe(self):
            print(f"Doctor - Name:{self.name} - Yob:{self.yob} - Specialist:{self.specialist}")
            
student1 = Student(name="studentA", yob=2010, grade="7")
teacher1 = Teacher(name="teacherA", yob=1969, subject="Math")
doctor1 = Doctor(name="doctorA", yob=1945, specialist="Endocrinologists")
teacher2 = Teacher(name="teacherB", yob=1995, subject="History")
doctor2 = Doctor(name="doctorB", yob=1975, specialist="Cardiologists")
ward1 = Ward(name="Ward1")
ward1.addPerson(student1)
ward1.addPerson(teacher1)
ward1.addPerson(teacher2)
ward1.addPerson(doctor1)
ward1.addPerson(doctor2)
ward1.sortAge()
ward1.describe()
ward1.aveTeacherYearOfBirth()
print(f"\nNumber of doctors: {ward1.countDoctor()}")
```

```python
#bai 3
class Stack():
    def __init__(self, capacity):
        self.limit = capacity
        self.stack = []
    def push(self,value):    
        if len(self.stack) <= self.limit:        
            self.stack.append(value)

    def isFull(self):
        if len(self.stack) == self.limit:
            return True
        else:
            return False
    def isEmpty(self):
        if len(self.stack) == 0:
            return True
        else:
            return False
    def pop(self):
        return self.stack.pop()
    def top(self):
        return self.stack[-1:][0]
        
stack1 = Stack(capacity=5)
stack1.push(1)
stack1.push(2)
stack1.push(4)
stack1.push(5)
stack1.push(3)
print(stack1.isFull())
print(stack1.top())
print(stack1.pop())
print(stack1.top())
print(stack1.pop())
print(stack1.isEmpty())
```

```python
#bai 4
class Queue():
    def __init__(self, capacity):
        self.limit = capacity
        self.queue = []
    def enqueue(self,value):    
        if len(self.queue) <= self.limit:        
            self.queue.append(value)
            
    def isFull(self):
        if len(self.queue) == self.limit:
            return True
        else:
            return False
    def isEmpty(self):
        if len(self.queue) == 0:
            return True
        else:
            return False
    def dequeue(self):
        return self.queue.pop(0)
    def front(self):
        return self.queue[:1][0]
    
queue1 = Queue(capacity=5)
queue1.enqueue(1)
queue1.enqueue(2)
print(queue1.isFull())
print(queue1.front())
print(queue1.dequeue())
print(queue1.front())
print(queue1.dequeue())
print(queue1.isEmpty())
```
